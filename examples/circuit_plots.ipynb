# Circuit plots for documentation

This notebook draws representative FP-QGPU circuits with Qiskit and exports PNG files used by Sphinx.

```python
from pathlib import Path
import matplotlib.pyplot as plt
from fp_qgpu.circuits import simple00, simple01, ghz_test
```

```python
def save_circuit_plot(qc, output_path: Path, title: str) -> None:
    fig = qc.draw(output='mpl')
    fig.suptitle(title)
    fig.savefig(output_path, dpi=180, bbox_inches='tight')
    plt.close(fig)
```

```python
output_dir = Path('../docs/_static')
output_dir.mkdir(parents=True, exist_ok=True)

save_circuit_plot(simple00(), output_dir / 'circuit_simple00.png', 'simple00()')
save_circuit_plot(simple01(), output_dir / 'circuit_simple01.png', 'simple01()')
save_circuit_plot(ghz_test(3), output_dir / 'circuit_ghz3.png', 'ghz_test(3)')

sorted(str(p.name) for p in output_dir.glob('circuit_*.png'))
```
